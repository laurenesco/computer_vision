```python
###
# Imports
###
import cv2
import numpy as np

###
# Step 1: Configure video capture
###
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    raise IOError("Cannot open webcam")

###
# Step 2: Configure window for output
###
WINDOW_NAME = "Canny Edge Detection (Live)"
cv2.namedWindow(WINDOW_NAME)

def nothing(x):
    # required callback for createTrackbar
    pass

cv2.createTrackbar("Low Threshold", WINDOW_NAME, 50, 500, nothing)
cv2.createTrackbar("High Threshold", WINDOW_NAME, 150, 500, nothing)

###
# Step 3: Main loop - read, preprocess, edge detect live
###
while True:
    # Iterate over video frames in real time
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break

    # Preprocess each frame (Canny requires one channel, benefits from Gaussian blur)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    # Read trackbar values
    low_thresh = cv2.getTrackbarPos("Low Threshold", WINDOW_NAME)
    high_thresh = cv2.getTrackbarPos("High Threshold", WINDOW_NAME)

    # Apply edge detection and display results
    edges = cv2.Canny(blurred, low_thresh, high_thresh)

    # Convert edges back to 3-channel so it can be stacked side-by-side with color frame
    edges_bgr = cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR)
    side_by_side = np.hstack((frame, edges_bgr))

    cv2.imshow(WINDOW_NAME, side_by_side)

    # Exit on 'q' keypress
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

###
# Step 4: Cleanup
###
cap.release()
cv2.destroyAllWindows()
```
